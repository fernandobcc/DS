```python
fig = plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='asiatico') #coloca legenda no grafico (plot x, y , label='legenda desejada')
ax.plot(x,x**3, label='africano')
ax.legend(loc=0) #loc=numero existe um inteiro para cada posição que a legenda pode ficar basta consultar o site do matplot para
#encontra a melhor posição por exemplo loc=0 - significa que o programa vai tentar achar a melhor loc. loc=1 no canto superior 
#direito. Ou voce pode usar um tuple para colocar exatamente onde quer. loc=(x,x)
ax.set_title('Penis')
ax.set_xlabel('Tamanho')
ax.set_ylabel('Ereção')
```

## AULA 3 APARENCIA

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple') #colocando a cor
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='#FF8C00') #com codigo RGB Hex
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='#FF8C00',lw=5,alpha=0.5) #linewidth (ou lw)=grossura da linha, alpha relaciona-se a transparencia
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='green',lw=5,linestyle='--') #linestyle (or ls)= como vai ser a linha (traços,linhas,ect.) 
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='green',lw=2,ls='-',marker='o',markersize=10,
       markerfacecolor='yellow',markeredgewidth=3,markeredgecolor='blue')
#marker marca os pontos do array e markersize=o tamanho do marker. Além disso a outras opções para customizar os marker.
#como exemplo a markerfacecolor='cor',markeredgewidth=expessura da linha,markeredgecolor='cor da linha', etc.
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='blue',lw=2,ls='--')
ax.set_xlim([0,1]) #insere um limite para o eixo x
ax.set_ylim([0,2]) #insere um limite para o eixo y
```
